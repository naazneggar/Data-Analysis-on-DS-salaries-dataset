# Comparing top 13 high salaried job title with experience levels

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))

# Sort and get the top 13 job titles by salary
sorted_data = data1.sort_values(by="salary_in_usd", ascending=False)
top10 = sorted_data.head(13)



# Plot
sns.barplot(
    x='salary_in_usd',
    y='job_title',
    data=top10,
    estimator='mean',
    hue='experience_level',
    
)

plt.xlabel('Average Salary (USD)')
plt.ylabel('Job Title')
plt.title('Top 13 Job Titles by Salary with Experience Levels')
plt.legend(title='Experience Level')
plt.show()
```

## It seems that mid level employees are paid more than senior level exployees in Research scientist and  Data analyst jobs. This seems counter intuitive
## May be its Performance-Based Pay: If mid-level employees are highly productive or contribute to critical projects, their performance-based bonuses MAY BE HIGH

## Or the data may be skewed,A few high salaries for mid-level roles could skew the average

# Lets compare maximum salary vs location

```python
plt.figure(figsize=(7,4))

data1.groupby("company_location")["salary_in_usd"].max().sort_values(ascending=False).head(5).plot(kind='pie',shadow=True,autopct="%1.2f%%")	

plt.title("MAximum Salary vs Location")


```

### Locations DE and IN has low maximum salaries and US has more salary

### Lets compare average salary vs location

```python
plt.figure(figsize=(7,4))

data1.groupby("company_location")["salary_in_usd"].mean().sort_values(ascending=False).head(5).plot(kind='pie',shadow=True,autopct="%1.2f%%")	

plt.title("Average Salary vs Location")


```

### Location IL has more average salary and US has less average salary related to Data science

# Lets compare average salary vs company size

```python
plt.figure(figsize=(7,4))

data1.groupby("company_size")["salary_in_usd"].mean().sort_values(ascending=False).plot(kind='pie',explode=[0.1,0,0],shadow=True,autopct="%1.2f%%")	

plt.title("Average Salary vs company size")


```

### The average salary in medium sized companies are more compared to small and large

```python
data2=pd.read_csv("ds_salaries.csv")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame creation


# Group by 'job_title' and 'experience_level', then calculate the mean salary
grouped_data = data2.groupby(['job_title', 'experience_level'])['salary_in_usd'].mean().reset_index()

# Sort values by 'salary_in_usd' in descending order
sorted_data = grouped_data.sort_values(by='salary_in_usd', ascending=False)
top10_jobs = sorted_data.head(10)

# Plot
plt.figure(figsize=(12, 8))
sns.barplot(
    x='salary_in_usd',
    y='job_title',
    hue='experience_level',
    data=top10_jobs,
    
)

plt.xlabel('Average Salary (USD)')
plt.ylabel('Job Title')
plt.title('Average Salary by Job Title and Experience Level')
plt.legend(title='Experience Level')
plt.show()
```
